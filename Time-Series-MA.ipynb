```python
# Time-Series Analysis: Moving Averages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a simple synthetic price series
np.random.seed(42)
days = 200
price = np.cumsum(np.random.randn(days)) + 100  # random walk around 100

# Convert to DataFrame
df = pd.DataFrame({
    "Day": range(days),
    "Price": price
})

df["MA20"] = df["Price"].rolling(window=20).mean()
df["MA50"] = df["Price"].rolling(window=50).mean()

# Plotting
plt.figure(figsize=(12,6))
plt.plot(df["Price"], label="Price", linewidth=1.2)
plt.plot(df["MA20"], label="20-Day MA")
plt.plot(df["MA50"], label="50-Day MA")

plt.title("Time-Series Trend: Price with Moving Averages")
plt.xlabel("Day")
plt.ylabel("Price Level")
plt.legend()
plt.grid(True)
plt.show()

df.tail()
```
